# Moving Average

```python
bank_stocks['Close']['TD']['2020'].plot()
```

```python
#Moving average for TD rolling 30 days average against the close price for these two banks for the year of 2020.
fig=plt.figure(figsize=(12,4))
ax = fig.add_axes([0,0,1,1])

ax.plot(bank_stocks['Close']['TD']['2020'].rolling(window=30).mean(), label="30 Days Average")
ax.plot(bank_stocks['Close']['TD']['2020'], label="RY Close")


ax.legend()
#Put a legend to the right of the plot
ax.legend(loc='center left', bbox_to_anchor=(1, 0.75))
```

```python
#Moving average for RY rolling 30 days average against the close price for these two banks for the year of 2020.
fig=plt.figure(figsize=(12,4))
ax = fig.add_axes([0,0,1,1])

ax.plot(bank_stocks['Close']['RY']['2020'].rolling(window=30).mean(), label="30 Days Average")
ax.plot(bank_stocks['Close']['RY']['2020'], label="RY Close")


ax.legend()
#Put a legend to the right of the plot
ax.legend(loc='center left', bbox_to_anchor=(1, 0.75))
```

```python
#Create a heatmap of the correlation between the stocks Close Price

sns.heatmap(bank_stocks.xs(key='Close',axis=1,level='Stock Info').corr(),annot=True,cmap='magma',linecolor='white',linewidths=1)
```

```python
# Create a clustermap of the correlation between the stocks Close Price

sns.clustermap(bank_stocks.xs(key='Close',axis=1,level='Stock Info').corr(),annot=True,cmap='magma',linecolor='white',linewidths=1)
```
